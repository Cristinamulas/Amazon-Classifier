# Web Scraping Amazon Webside

-  I scraped Amazon webside to get the reviews from an APP called OverDrive that is a librery for audio books and Ebooks. I wrote a scrip that internate over 200 pages for the Amazon site and extract the title, review, person who reviewed and the date when the review took place from this App.

```python
# importing usefull libreries
import requests
import urllib.parse
import urllib.error
from bs4 import BeautifulSoup
import pandas as pd
import ssl
from functions import *
```

##  Web Scraper Pagination Script

```python
reviews = []
dates = []
names = []
tittles = []
for i in range(1,200):
    ctx = ssl.create_default_context()
    ctx.check_hostname = False
    ctx.verify_mode = ssl.CERT_NONE
    if i == 1:
        url = 'https://www.amazon.com/product-reviews/B00850NK6I/ref=cm_cr_getr_d_paging_btm_prev_1?ie=UTF8&filterByStar=critical&reviewerType=all_reviews&pageNumber=1'
    else:
         url ="https://www.amazon.com/product-reviews/B00850NK6I/ref=cm_cr_getr_d_paging_btm_prev_1?ie=UTF8&filterByStar=critical&reviewerType=all_reviews&pageNumber={}".format(i)
    html = urllib.request.urlopen(url, context=ctx).read()
    soup = BeautifulSoup(html, 'html.parser')
    reviews += getting_reviews(soup)
    dates+= getting_dates(soup)
    names += getting_names(soup)
    tittles += getting_tittles(soup)
df = pd.DataFrame([reviews, dates, names, tittles]).transpose()
df.columns = ['Reviews', 'Date', 'Names', 'Tittles']
print(len(df))
df
```

```python
# save as csv file
df.to_csv("reviews.csv" , index = False)
```
